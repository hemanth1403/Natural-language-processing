a. How many words are there in the given text file

```python
n = input("Enter file name : ")
f = open(n,"r")
c = 0
for i in f:
  a = i.split()
  c += len(a)
print(c)
```

---------------------------------------------------------
b. How many different words are there in the given text 

```python
l = input("Enter the statement : ").split()
a = {}
for i in l:
  if i not in a:
    a[i] = 1
  else:
    a[i] += 1
print(a)
```

c. How many times does the word occur in the text file

```python
n = input("Enter the file name : ")
f = open(n,"r")
s = input("Enter substring : ")
c = 0
for i in f:
  a = i.split()
  c += a.count(s)
print(s,":",c)
```

d. What is the percentage of all the words in the text file

```python
def percentage(n, total):
  d = {}
  for i in range(len(n)):
    for j in n[i]:
      if(j not in d):
        d[j] = 1
      else:
        d[j] += 1
  temp = list(d.keys())
  temp = sorted(temp)
  for i in temp:
    print(i,":", (d[i]*100)/total,"%")
```

```python
ls = []
total = 0
for i in range(int(input("Enter no of statements : "))):
  n = input("Enter statement : ").split()
  total += len(n)
  ls.append(n)
percentage(ls, total)
```
