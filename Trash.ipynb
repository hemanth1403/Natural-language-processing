```python
def solution(ls):
  d = {}
  for i in range(len(ls)):
    for j in ls[i].split():
      if(j not in d):
        d[j] = [i+1]
      else:
        d[j].append(i+1)
  print(d)
ls = ["My name is", "name is", "name"]
solution(ls)
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
wt = word_tokenize("This is a sample text")
wt
```

```python
# In this the sentence can only be tokentized by the fullstop followed by NOTHING or New sentence followed by a new sentence with a capital letter
st = sent_tokenize("this is a.   Sample sentence.")
st
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
sentence = "This is a sample sentence"
tokens = word_tokenize(sentence)
ps = PorterStemmer()
wnl = WordNetLemmatizer()
for i in tokens:
  print("Stemmer : ", ps.stem(i),end=",   ")
  print("Lemmatize : ", wnl.lemmatize(i))
```
