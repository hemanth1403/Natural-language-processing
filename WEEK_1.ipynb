```python
def solution(ls):
  d = {}
  for i in range(len(ls)):
    for j in ls[i].split():
      if(j not in d):
        d[j] = [i+1]
      else:
        d[j].append(i+1)
  for i,j in d.items():
    print(i,":",j)
```

```python
ls = []
for i in range(int(input("Enter no of statements : "))):
  ls.append(input("Enter statement : "))
solution(ls)
```
